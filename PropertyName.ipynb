I used this pyhton script to scrap data from brickz.my website fro my research project. The main objectives from this scrapping 
is to get the name of the property and all the transactions number from january 2008 until fab 2018. All the name of the property will then be  used to scrap all the transaction for a particular property

```python
# import all package
from bs4 import BeautifulSoup
from requests import get
import requests
#headers = {'User-Agent': 'Mozilla/5.0 (X11; U; Linux i686; en-US) AppleWebKit/534.3 (KHTML, like Gecko) Chrome/6.0.472.63 Safari/534.3'}
headers = {'User-Agent': 'Mozilla/5.0 (Linux; Android 5.1.1; SM-G928X Build/LMY47X) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.83 Mobile Safari/537.36'}
result = requests.get("http://derstandard.at", headers=headers)
c = result.content
print (result.request.headers)


print (len(c))
```

```python

pages = [str(i) for i in range(1,107)]

names = []
areas = []
tenures = []
type_res = []
median_Prices = []
transactions = []

for page in pages:
    
    response = get("https://www.brickz.my/transactions/residential/kuala-lumpur/non-landed/page/"+ page + "/?range=2008+Jan-", headers = headers)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    residential_container = html_soup.find_all("tr", itemtype="https://schema.org/AdministrativeArea")
    print(response)
    
#Extract data from individual residential_container
    for container in residential_container:
        #The Name 
        name = container.td.find("span",itemprop="name" ).text
        names.append(name)

        #The area
        area = container.td.find_all("span" )[1].find("span",itemprop="name" ).text
        areas.append(area)

        #The Tenures
        tenure = container.find("span", class_="ptd_list_item_title").text
        tenures.append(tenure)

        #The Types
        type_re = container.find_all("td")[1].text
        type_res.append(type_re)

        #The Median Price
        median_Price = container.find_all("span", class_="ptd_list_item_title ptd_currency")[0].text
        median_Prices.append(median_Price)
        
        #Number of transaction
        transaction = container.find("a", class_="button").text
        transactions.append(transaction)
```

```python
import pandas as pd
Data = pd.DataFrame({"Name":names,
                       "Street":areas,
                       "Tenure":tenures,
                       "Type":type_res,
                       "Median Price Psf":median_Prices,
                       "No of Transactions": transactions}
                      )
```
