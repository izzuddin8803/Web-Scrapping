```python
import pandas as pd
import numpy as np
#Upload the residential.csv to get the property name
Residential = pd.read_csv("Residential.csv")
Residential.drop(["Unnamed: 0","Unnamed: 7"], axis=1, inplace=True)
```

```python
Residential[1:4]
```

```python
Test=Residential[0:]
```

```python
import googlemaps
from datetime import datetime

gmaps = googlemaps.Client(key="The Key")
```

```python
Geocode2 =[]

#For loop our python script, we put kuala lumpur to make sure the google map find coordinate inside kuala lumpur
for name in Test.Name:
    try:
        geocode_result = gmaps.geocode(name + "Kuala Lumpur")[0]["geometry"]["location"]
        print(geocode_result)
        geocode_result["Name"]=name
        Geocode2.append(geocode_result)
    except IndexError:
         #if the no result found
        geocode_result={'lat': "Missing", 'lng': "Missing"}
        print(geocode_result)
        geocode_result["Name"]=name
        Geocode2.append(geocode_result)
```
